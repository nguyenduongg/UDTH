```python
from vnstock import Quote
import time
```

```python
import os

ticker_list = ["VCB", "FPT", "MWG", "VNM", "HPG", "TCH", "CTD", "SSI", "DGC", "PLX"]
start_date = "2020-01-01"
end_date = "2026-04-16"

for ticker in ticker_list:
    try:
        quote = Quote(symbol=ticker, source='VCI')
        df = quote.history(start=start_date,end=end_date,interval='d')

        if df is not None and not df.empty:
            file_path = os.path.join('Data',f"stock_data_{ticker}.csv")
            df.to_csv(file_path, index = False)

            print('Đã lưu: {file_path}')
            time.sleep(3)
    except Exception as e:
        print(f"Lỗi mã {ticker}:{e}")
```
